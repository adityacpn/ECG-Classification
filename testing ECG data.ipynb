```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('/home/aditya/ML_prog/DATA/Arrythmia database/archive/100.csv')
data1.head()
#data1.shape[1]
```

```python
#plt.figure(figsize(50,10))
f = plt.figure()
f.set_figwidth(16)
f.set_figheight(6)

plt.plot(data1['MLII'][:4000], color='r', label='MLII')
plt.plot(data1['V5'][:4000], color='g', label='V5')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Sample")
plt.ylabel("Readings")
plt.title("ECG Reading on V5 and MLII")

# Adding legend, which helps us recognize the curve according to it's color
plt.legend()

# To load the display window
plt.show()
```
